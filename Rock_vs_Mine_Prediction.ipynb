Project Description :- The "Rock vs. Mine Prediction" project focuses on predicting whether an underwater object is a rock or a mine using machine learning techniques.

Goal :- The goal of the "Rock vs. Mine Prediction" project is to develop a model that can accurately classify underwater objects as either rocks or mines. This classification task has significant applications in areas such as marine exploration and defense.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and processing

```python
sonar_data = pd.read_csv('/content/Copy of sonar data.csv', header = None)
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
#sepating features and label
X = sonar_data.drop(columns = 60, axis = 1)
y = sonar_data[60]
```

```python
print(X)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, stratify = y, random_state = 1)
```

```python
print(X.shape)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
print(x_train)
print(y_train)
```

Model Training --> Logistic Regression

```python
model = LogisticRegression()
```

```python
 model.fit(x_train, y_train)
```

Model Evaluation

```python
#accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print(test_data_accuracy)
```
